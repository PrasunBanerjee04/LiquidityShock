```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/training.csv')
df
```

```python
#visualize the price time series 
filtered_df = df[(df['security_id'] == 1) & (df.index <= 100000)]

plt.figure(figsize=(12,6))
plt.plot(filtered_df.index, filtered_df['ask1'])
plt.xlabel("Index")
plt.ylabel("Price")
plt.title('Price Series Visualization')
plt.grid(True)
plt.show()
```
